```python
Order_Details.dropna(subset=['Order ID', 'User ID'], inplace=True)
```

```python
User_Details['Registration Date'] = pd.to_datetime(User_Details['Registration Date'])
Cooking_Sessions['Session Start'] = pd.to_datetime(Cooking_Sessions['Session Start'])
Cooking_Sessions['Session End'] = pd.to_datetime(Cooking_Sessions['Session End'])
Order_Details['Order Date'] = pd.to_datetime(Order_Details['Order Date'])
```

```python
user_details = User_Details.drop_duplicates()
cooking_sessions = Cooking_Sessions.drop_duplicates()
order_details = Order_Details.drop_duplicates()
```

```python
user_sessions = pd.merge(user_details, cooking_sessions, on='User ID', how='inner')
```

```python
full_data = pd.merge(user_sessions, order_details, on=['User ID', 'Session ID'], how='inner')
```

```python
full_data.head()
```

```python
# Correlation between Session Rating and Order Rating
correlation = full_data['Session Rating'].corr(full_data['Rating'])
print(f"Correlation between session and order ratings: {correlation:.2f}")
```

```python
# Count the frequency of each dish
popular_dishes = full_data['Dish Name_y'].value_counts()
print("Popular Dishes:\n", popular_dishes)
```

```python
# Average orders by age group
age_groups = pd.cut(user_details['Age'], bins=[0, 18, 30, 45, 60, 100], labels=['<18', '18-30', '30-45', '45-60', '60+'])
user_details['Age Group'] = age_groups

orders_by_age_group = user_details.groupby('Age Group')['Total Orders'].mean()
print("Average Orders by Age Group:\n", orders_by_age_group)
```

```python
# Total and average revenue
total_revenue = full_data['Amount (USD)'].sum()
average_order_value = full_data['Amount (USD)'].mean()
print(f"Total Revenue: ${total_revenue:.2f}")
print(f"Average Order Value: ${average_order_value:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
popular_dishes.plot(kind='bar', color='skyblue', title='Most Popular Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Age Group Distribution
user_details['Age Group'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='User Age Group Distribution')
plt.ylabel('')
plt.show()
```

```python
sns.barplot(data=full_data, x='Session Rating', y='Rating')
plt.title('Session Rating V/S Order Rating')
plt.xlabel('Session Rating')
plt.ylabel('Order Rating')
plt.show()
```
